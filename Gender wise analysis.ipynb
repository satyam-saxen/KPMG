```python
#sns.barplot(x="gender", y="past_3_years_bike_related_purchases", data=df)
male_c=0
female_c=0
u_c=0
male_count=0
female_count=0
u_count=0
for index,rows in df.iterrows():
    if rows['gender']== 'Male':
        male_c += int(rows['past_3_years_bike_related_purchases'])
        male_count += 1
    elif rows['gender'] == 'Female':
        female_c += int(rows['past_3_years_bike_related_purchases'])
        female_count +=1
    else:
        u_count += 1
        u_c += int(rows['past_3_years_bike_related_purchases'])
```

```python
data = {'Gender':['Male','Male_P','Female','Female_P','U','U_P'],
       'BikeP':[male_count,male_c,female_count,female_c,u_count,u_c]}
```
